```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import LogLocator, FuncFormatter

g2 = 0.652
LeudH7_dLueH_factor = 2

# Data of D7RGESolver at 10 TeV

solve_rge_D7RGESolver = [
    
    (1.922e-20)*1e12, 
    (1.874e-16)*1e12,  
    (1.067e-18)*1e12, 
    (9.992e-19)*1e12,    
    (4.256e-19)*1e12,  
    (6.383e-17)*1e12,  
    (2.676e-16)*1e12,  
    (2.179e-14)*1e12,
    (2.775e-17)*1e12, 
    (1.918e-15)*1e12,
    
]


# Data of nudobe at 10 TeV
solve_rge_nudobe = [
    
    (1.651e-20)*1e12, 
    (1.917e-16)*1e12, 
    (4.506e-12)*1e12,  
    (8.222e-12)*1e12, 
    (1.029e-12)*1e12*g2,  
    (6.249e-17)*1e12,  
    (2.612e-16)*1e12, 
    (1.09e-14*LeudH7_dLueH_factor)*1e12, 
    (2.789e-17)*1e12,  
    (1.872e-15)*1e12, 
    
]


# Data of electroweak scale: no RG running
EW_values = [
      
    1.651e-20*1e12,
    1.517e-16*1e12,  
    3.361e-12*1e12,  
    6.109e-12*1e12,    
    5.636e-13*1e12,   
    7.987e-17*1e12,  
    3.311e-16*1e12, 
    2.138e-14*1e12, 
    3.471e-17*1e12,  
    1.881e-15*1e12, 
]

operators = [
    r"$C_{LH}^{11}$", r"$C_{LeDH}^{11}$", r"$C_{DLDH1}^{11}$", r"$C_{DLDH2}^{11}$", 
     r"$C_{LHW}^{11}$",  r"$C_{\overline{d}LQLH1}^{1111}$", 
    r"$C_{\overline{d}LQLH2}^{1111}$", r"$C_{\overline{d}LueH}^{1111}$", r"$C_{\overline{Q}uLLH}^{1111}$",r"$C_{\overline{d}uLDL}^{1111}$"
]


fig, ax = plt.subplots(figsize=(14, 6))
x = np.arange(len(operators))
width = 0.25

# Bar plots
ax.bar(x - width, solve_rge_D7RGESolver, width, label="With RG running, D7RGESolver", color="tab:blue", zorder=2)
ax.bar(x, solve_rge_nudobe, width, label=r"With RG running, $\nu \text{DoBe}$", color="tab:orange", zorder=2)
ax.bar(x + width, EW_values, width, label="Without RG running", color="tab:green", zorder=2)
  
ax.set_ylabel(r"$|c_7^i|\equiv |\Lambda^{3}C_7^i(\Lambda)|$", fontsize=28) 
ax.set_yscale("log")
ax.set_ylim(1e-8, 1e2)


ax.yaxis.set_major_locator(LogLocator(numticks=15))
ax.yaxis.set_minor_locator(LogLocator(subs='all', numticks=15))
ax.tick_params(axis='y', which='both',      direction='in',    length=3,         labelsize=18,      right=True         )
ax.grid(which='major', axis='y', linestyle='--', linewidth=1, color='gray', alpha=1)


ax.axhline(1e-7, linestyle='--', linewidth=1, color='gray', alpha=1, zorder=1)
ax.axhline(1e-5, linestyle='--', linewidth=1, color='gray', alpha=1, zorder=1)
ax.axhline(1e-3, linestyle='--', linewidth=1, color='gray', alpha=1, zorder=1)
ax.axhline(1e-1, linestyle='--', linewidth=1, color='gray', alpha=1, zorder=1)
ax.axhline(1e1, linestyle='--', linewidth=1, color='gray', alpha=1, zorder=1)


ax.set_xticks(x)
ax.set_xticklabels(operators, fontsize=22)
ax.tick_params(axis='x', pad=20)


ax.text(
    0.02, 0.89, r"$\Lambda=10$ TeV",
    fontsize=30,
    transform=ax.transAxes,
    bbox=dict(facecolor='white', edgecolor='gray', alpha=1, boxstyle='round,pad=0.3')
)

legend = ax.legend(fontsize=20, bbox_to_anchor=(1.005, 1.01), loc='upper right')
legend.get_frame().set_alpha(1)  
ax.tick_params(axis='both', which='both', direction='in', right=True)




plt.tight_layout()
plt.savefig("Figures/compare_run_three.pdf", format='pdf')
plt.show()
```
