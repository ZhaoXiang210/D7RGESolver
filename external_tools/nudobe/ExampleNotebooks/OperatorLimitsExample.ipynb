# Get Limits from a Model Class

Getting limits from a model class can be more useful than getting it from the functions module if you want to change PSFs, LECs or NMEs manually.

```python
import numpy as np
```

```python
#Initiate models
LEFT_model = nudobe.EFT.LEFT({}, method = "SM", unknown_LECs = True)
SMEFT_model = nudobe.EFT.SMEFT({}, method = "SM", unknown_LECs = True)
```

```python
#Generate limits
LEFT_lims  = LEFT_model.get_limits(2.3e+26, "136Xe", groups = True)
SMEFT_lims = SMEFT_model.get_limits(2.3e+26, "136Xe")
```

```python
LEFT_lims
```

```python
SMEFT_lims
```

```python
#randomly change unknown LECs
for LEC in nudobe.constants.LECs_unknown:
    #draw from a normal distribution
    rand = np.random.normal(1, 0.5)
    
    #randomize the sign except for g_nu^NN
    if LEC != "nuNN":
        sign = np.random.choice([-1, 1])
    else:
        sign = 1
    LEFT_model.LEC[LEC]  *= sign*rand
    SMEFT_model.LEC[LEC] *= sign*rand
```

```python
#Get Limits after randomizing LECs
LEFT_lims_rand  = LEFT_model.get_limits(2.3e+26, "136Xe", groups = True)
SMEFT_lims_rand = SMEFT_model.get_limits(2.3e+26, "136Xe")
```

```python
#Difference in limits
(LEFT_lims - LEFT_lims_rand)/LEFT_lims
```

```python
(SMEFT_lims - SMEFT_lims_rand)/SMEFT_lims
```
