# Compare the RGEs of $0\nu\beta\beta$ operators solved by D7RGESolver with $\nu \text{DoBe}$

## LH5_11 vs LH(5)

### RGE solved by D7RGESolver

```python
C_in = {"LH5_11": 1e-15+0j}       # Input the WCs
scale_in = 1e4                   # Input energy scale in unit of GeV
scale_out = 80                   # Output energy scale in unit of GeV
C_out = solve_rge(scale_in, scale_out, C_in, basis="down", method="integrate")  
print_WCs(C_out)
```

### RGE solved by $\nu \text{DoBe}$

```python
C_in_Ops="LH(5)"
scale_in=1e4
C_in_ref = {C_in_Ops: 1e-15}
model = SMEFT(C_in_ref)   
C_out_0vbb_ref=model.run(initial_scale = scale_in, inplace = True) ## The default final_scale is EW scale=80 GeV
C_out_0vbb_ref
```
