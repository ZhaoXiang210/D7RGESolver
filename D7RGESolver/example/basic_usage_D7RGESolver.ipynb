## Import stuff of D7RGESolver

```python
import numpy as np
import sys
import os
sys.path.append("..")  
import matplotlib.pyplot as plt
from RGE_dim7 import solve_rge, print_WCs,plot_WC
%matplotlib inline
```

## Solve RGEs using D7RGESolver

#### Choose the down-quark flavor basis and accurate integrated method (funtion solve_reg) to solve the RGEs

```python
C_in = {"LH5_11": 1e-15+0j,"DLDH1_11": 1e-15+0j}       # Input the WCs
scale_in = 1e4                   # Input energy scale in unit of GeV
scale_out = 80                   # Output energy scale in unit of GeV
C_out = solve_rge(scale_in, scale_out, C_in, basis="down", method="integrate")  
print_WCs(C_out)
```

### Print the specific WCs in C_out

```python
print_WCs(C_out,specific_keys=["LH5_11","LH_11"])
```

#### Choose the up-quark flavor basis and leadinglog approxiamtion to solve the RGEs

```python
C_in = {"LH5_11": 1e-15+0j,"DLDH1_11": 1e-15+0j}       
scale_in = 1e4                   
scale_out = 80                  
C_out = solve_rge(scale_in, scale_out, C_in, basis="up", method="leadinglog")  
print_WCs(C_out)
```

### Plot the RG running of the WC itself

```python
plot_WC(
    scale_initial=80, 
    scale_final=1e10, 
    C_in_initial={"LH5_11": 1e-15},  
    C_yaxis={"LH5_11"},
    number_points=10, 
    method='integrate', 
    basis="down",
    xscale='log', 
    yscale='log', 
    rtol=1e-6, 
    atol=1e-8, 
    mode='abs'
)
```

### Plot the RG running of the SM paramater

```python
plot_WC(
    scale_initial=1e2, 
    scale_final=1e20,  
    C_in_initial={},  
    C_yaxis={"gp", "g", "gs"},
    number_points=15, 
    method='integrate', 
    basis="down",
    legend_labels=None, 
    xlabel=None, 
    ylabel=r"SM parameters", 
    caption=r"RG evolution of SM parameters", 
    xscale='log', 
    yscale='linear', 
    rtol=1e-6, 
    atol=1e-8, 
    mode='abs'
)
```
